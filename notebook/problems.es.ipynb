## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
is_par = 0
is_greater_than_7 = 0

def throw_dice(): 
    return np.random.randint(1,7)

def get_throws_sum():
    dice_a = throw_dice()
    dice_b = throw_dice()

    return dice_a + dice_b

def get_probability(n, i):
    probability = n / i
    return round(probability * 100, 2)

for i in range(1000):
    result = get_throws_sum()

    if result % 2 == 0:
        is_par += 1
    
    if result > 7:
        is_greater_than_7 +=1

print(f"Veces que hemos obtenido números pares: {is_par}")
print(f"Probabilidad de obtener un número par: {get_probability(is_par, 1000)}%")
print(f"Veces que hemos obtenido números mayores que 7: {is_greater_than_7}")
print(f"Probabilidad de obtener un número mayor que 7: {get_probability(is_greater_than_7, 1000)}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
is_3_white_2_red = 0
is_same_color = 0

def take_a_ball(box):
    return np.random.randint(1, len(box))

def get_balls_colors(box):
    is_white = 0
    is_red = 0
    is_green = 0

    for i in range(5):
        ball = take_a_ball(box)
        ball_color = ball_box[ball]
        
        if ball_color == 'White':
            is_white += 1
        elif ball_color == 'Red':
            is_red += 1
        elif ball_color == 'Green':
            is_green += 1

    return {'White': is_white, 'Red': is_red, 'Green': is_green}

def get_probability(n, i):
    probability = n / i
    return round(probability * 100, 2)

for i in range(1000):
    balls_colors = get_balls_colors(ball_box)

    if balls_colors['White'] == 3 and balls_colors['Red'] == 2:
        is_3_white_2_red += 1
    
    if balls_colors['White'] == 5 or balls_colors['Red'] == 5 or balls_colors['Green'] == 5:
        is_same_color +=1

print(f"Veces que han salido tres blancas y dos rojas: {is_3_white_2_red}")
print(f"Probabilidad de obtener tres blancas y dos rojas por tirada: {get_probability(is_3_white_2_red, 1000)}%")
print(f"Veces que han salido todas del mismo color: {is_same_color}")
print(f"Probabilidad de obtener todas del mismo color por tirada: {get_probability(is_same_color, 1000)}%")
```
